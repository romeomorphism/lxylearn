```python
from sklearn.cluster import SpectralClustering as SpectralClustering_sklearn
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np

n_samples = 500
noise_circle = datasets.make_circles(
    n_samples=n_samples, factor=0.6, noise=0.05
)

center = [[0,0]]
data1, labels = noise_circle
data2, labels2 = datasets.make_blobs(n_samples=100, 
                            cluster_std = 0.1,
                            centers=center,
                            random_state=1)

labels2 = np.ones_like(labels2) * 3
labels = np.concatenate([labels, labels2])
data = np.concatenate([data1, data2], axis=0)

plt.scatter(data[:, 0], data[:, 1], c=labels)
```

```python
from spectral import SpectralClustering

spectral = SpectralClustering(n_clusters=3, affinity='nearest_neighbors', n_neighbors=15)
spectral.fit(data, normalized=True)

```

```python
plt.figure(figsize=(6, 3))
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1])
plt.subplot(1, 2, 2)
plt.scatter(data[:, 0], data[:, 1], c=spectral.labels_)
plt.title('Spectral Clustering')
plt.tight_layout()
plt.savefig('three clustering.pdf')
```

```python
from mpl_toolkits.mplot3d import Axes3D

print(spectral.U_.shape)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(spectral.U_[:, 0], spectral.U_[:, 1], spectral.U_[:, 2], c=spectral.labels_, rasterized=True)
ax.set_title('3D Scatter plot of U')
plt.savefig('projected data.pdf')
plt.show()
```
