```python

from sklearn.cluster import SpectralClustering as SpectralClustering_sklearn
from sklearn import datasets
import matplotlib.pyplot as plt

n_samples = 500
noise_circle = datasets.make_circles(
    n_samples=n_samples, factor=0.5, noise=0.05
)

X, y = noise_circle
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
from spectral import SpectralClustering
from kmeans import KMeans

kmeans = KMeans(n_clusters=2)
spectral = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', n_neighbors=10)
# spectral_sklearn = SpectralClustering_sklearn(n_clusters=2, affinity='nearest_neighbors', n_neighbors=10)

spectral.fit(X)
# spectral_sklearn.fit(X)

kmeans.fit(X)
```

```python
plt.figure(figsize=(6, 3))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
plt.title('KMeans(Lloyd)')
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=spectral.labels_)
plt.title('Spectral Clustering')
plt.tight_layout()
plt.savefig('nonlinear clustering.pdf')


```
